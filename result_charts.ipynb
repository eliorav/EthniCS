```python
from pathlib import Path
import pandas as pd
from EthniCS.services.graphics.render_experiments_charts import *
from EthniCS.constants import ethnicities_to_full_name_mapping
```

```python
stats_folder = Path("./data/stats")
n=1024
ethnicity_names = list(ethnicities_to_full_name_mapping.keys())
```

## ALGN OVER AFR


```python
super_population_stats_df = pd.read_csv(stats_folder/'super_population_stats.csv')
```

```python
render_by_sparse_level(super_population_stats_df, metric='mse', non_zero1=0.01, non_zero2=0.07, num_of_pools=256, log_y1=True, log_y2=False, n=n)
```

```python
render_by_sparse_level(super_population_stats_df, metric='psnr', non_zero1=0.01, non_zero2=0.07, num_of_pools=256, log_y1=False, log_y2=False, n=n)
```

```python
render_grid_error_bar_by_non_zero(super_population_stats_df, title_text='PSNR error bar by k over AFR')
```

```python
render_grid_error_bar_by_num_of_pools(super_population_stats_df, range=np.arange(0.01, 0.11, 0.01), title_text='PSNR error bar by m over AFR')
```
